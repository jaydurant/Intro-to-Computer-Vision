The green monochromatic image more accurately displays the image as expected. I would expect that the green monochromatic image to be work better as in a computer vision algorithm as there is more of this color represented in this image and as such this intensity from this color more accurately represents the original images light intensity.


```
import math

total_px = 100
height, width = img1_green.shape
mid_h = height/2 - 1
mid_w = width/2 - 1

half_l = math.floor((total_px - 1) / 2)
half_u = total_px - 1 - half_l

begin_row = int(mid_h - half_l)
last_row = int(begin_row + total_px)

begin_col = int(mid_w - half_l)
last_col = int(begin_col + total_px)

center = img1_green[begin_row:last_row, begin_col:last_col]

# green image
img2_green = img2[:,:,1]

height, width = img2_green.shape  
mid_h = height / 2
mid_w = width / 2

begin_row = int(mid_h - half_l)
last_row = begin_row + total_px
begin_col = int(mid_w - half_l)
last_col = begin_col + total_px

img2_green[begin_row:last_row,begin_col:last_col] = center

cv2_imshow(img2_green)
cv.imwrite('output/ps0-3-a-1.png', img2_green)
```

```
min_green = np.amin(img1_green)
max_green = np.amax(img1_green)
mean_green = np.mean(img1_green)
std_green = np.std(img1_green)

print("Min {} Max {} Mean {} STD {}".format(min_green, max_green, mean_green, std_green))

new_green = ((img1_green - mean_green) / std_green) * 10 + mean_green

cv2_imshow(new_green)
cv.imwrite('output/ps0-4-b-1.png', new_green)
```

```
img1_green_left = np.roll(img1_green,-2, axis=1)
cv2_imshow(img1_green_left)
cv.imwrite('output/ps0-4-c-1.png', img1_green_left)
```

```
diff_img = np.abs(img1_green - img1_green_left)
cv.imwrite('output/ps0-4-d-1.png', diff_img)
cv2_imshow(diff_img)
```

Negative values indicate a difference outside of the range that can be represented. It means the image value being deducted from the original value is much higher or lighter than the previous value.

```
height, width, _ = img1.shape
normal_dist = np.random.normal(0, 25, size=(height, width))

img1[:,:,1] = img1[:,:,1] + normal_dist

cv2_imshow(img1)
cv.imwrite('output/ps0-5-a-1.png', img1)
```

```
img1[:,:,1] = img1[:,:,1] - normal_dist
img1[:,:,0] = img1[:,:,0] + normal_dist

cv2_imshow(img1)
cv.imwrite('output/ps0-5-b-1.png', img1)
```

I needed to use sigma 10 to have the noise become more visible.This sigma appears to increase the value of the green intensity.

The noise being added to the green channel has the greatest visible effect as the image contains higher amounts of green intensities across all pixels in the image and as those image channel values are changed there is a greater visible effect on the image as a whole. As such there is a greater and more visible effect when noise effects the green channel.
